**Abstract :** Groundwater Research is vital to resource management and long-term planning in Africa. However, the level of research output, and mode of  North-South research collaboration in the continent has been questioned. The term `Helicopter research` has been documented in various studies such as Biology, Geosciences and Soil science. However, its presence in Groundwater Research in Africa is unknown. This study embarked on a bibliometric study to review the publications of groundwater research in the continent by extracting and analyzing meta data of `5728 articles` from the `Clarivate “Web of Science. © Copyright Clarivate 2020”` database. Analysis of the results showed that there is an intra-continental imbalance in research distribution. 51.6% of the total groundwater research output in Africa was produced by `Egypt, Nigeria, Tunisia, Republic of South Africa (R.S.A) and Morocco`. Compared to the USA, `Africa produced only 34% more research`, despite being 3 times bigger in space and about 3.7 times more populated. Even though there was a relatively high amount of research partnership between the Global North and South, there was a large intra-continental imbalance.  A rather concerning trend is that African countries barely spearheads their research, with only `57% of the total research output authored locally`. By forming an index to investigate Helicopter research, it was discovered that `37 out of 55 African countries experienced Helicopter research`. The Research output can also be influenced by several factors, these include Hydrogeology and climate, addressing a country’s water needs, level of infrastructure, level of investment, Geographic context, and political will. To foster groundwater research, African Governments should invest in its infrastructure, pay attention to retaining its talent and ensure mutual collaboration with International Research Institutions.

### About the Dataset

**Processed Data :**
* `Local` in context to this study refers to an affiliation to a target country’s institution. Study period refers to the span of our search. In this study the study period was from 1991-2020 which is a total of 30 years
* `African author` is defined as a researcher who is affiliated to an African Institution; hence the term African Author is rather a geographic concept rather than ethnical. 
* `n-articles` or `Research Output (RO)` is the total number of articles accounted for within the study period.
* `Local co-authorship` is the total number of articles including at least one author from the target country.
* `African co-authorship` is the total number of articles including at least one African Author.
* `Local corresponding authorship` is the total number of articles authored by a target country's researcher
___

* `CLASS` is a distribution made based on `n-articles`

|Class||Class Int (RO)||RO Class||%RO class||Sub-Classes||Sub-classes Int||RO sub-class||% RO sub-class||No. of countries|

|-||-||-||-||B2||11 - 50||353||6.2||14|
|C||< 10||83||1.5||C||< 10||83||1.5||18|
___

* `Helicopter research index (HRI)` : is defined as the `n-articles without local co-authorship (NWLCA)` divided by `n-articles (NA)` of a given country

$$HRI = N_WLCA/N_A$$

**World Bank Statistics Data :** `Inhabitants per groundwater research`, `Population 2020 (million)`, `Area (km2)`, `Water Stress Index`, `GDP (Millions USD)`, `Research as a % of GDP`.

## 7. Conclusion

    North et. Al’s (2020) research “out of Africa” argued that there is a lack of North-South research partnership in high end research in Geoscience with low representation of African researchers in high impact Geoscience publications. This study showed that the latter, which is Helicopter Research is also present in groundwater research in Africa. It was observed that total groundwater research output in Africa is relatively low, heavily dictated by a few countries, most of which are in the Northern part of the continent. It also answered the question of the presence of Helicopter research between Global North-South research partnership. Although there is a role to be played by developed countries, developing countries and their respective scientific communities, the main responsibility lies on Governments of developing countries as they are mandated to meet growing water needs of their countries. The first step Governments should take towards achieving better groundwater research, is by investing in their country’s research infrastructure and create the platform for local researchers which will foster and strengthen future research collaboration.
